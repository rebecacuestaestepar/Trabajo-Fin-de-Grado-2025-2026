### Openpyxl

Primero vamos a trastear con openpyxl para ver cómo puedo sacar los datos del excel

```python
from openpyxl import load_workbook

wb = load_workbook("HORARIOS VENA(Curso 24-25).xlsx")

# Sacamos los nombres de las hojas que hay en el excel
print(wb.sheetnames)
```

```python
# Analizamos el horario para el grado en ingeniería informática
gii=wb.worksheets[5]

for row in gii.iter_rows():
    for cell in row:
        print(cell.coordinate, cell.value)
```

```python
for row in gii.iter_rows(values_only=True):
    print(row)
```

```python
for merged in gii.merged_cells.ranges:
    print(merged)

# Me devuelve el rango de las celdas mergeadas pero no en orden
```

```python
cell=gii["B5"]
print(cell.fill.start_color.rgb)

# El color no es un string, y por tanto, no devuelve el color de la celda
```

```python
cell2=gii["B6"]
cell2.alignment.textRotation

# La rotación me la devuelve de manera correcta
# 0 -> Texto horizontal
# 90 -> Texto vertical
```

```python
"""
Ordenamos las celdas mergeadas, para poder extraer bien los datos
"""

from openpyxl.utils.cell import coordinate_from_string, column_index_from_string

def sort_key_range(rng):
    start = str(rng).split(":")[0]  
    col_letters, row = coordinate_from_string(start) 
    return (row, column_index_from_string(col_letters))

merged_sorted = sorted(gii.merged_cells.ranges, key=sort_key_range)

for r in merged_sorted:
    print(r)
```

```python
# Quiero saber exactamente qué tipo es el color que hay en el excel
# Investigamos los diferentes tipos para verqué me devuelve para cada uno de ellos

c = cell.fill.start_color
print("type:", c.type)         
print("rgb:", c.rgb)
print("theme:", c.theme)
print("indexed:", c.indexed)
```

```python
c2 = cell2.fill.start_color
print("type:", c2.type)         
print("rgb:", c2.rgb)
print("theme:", c2.theme)
print("indexed:", c2.indexed)

# El color es un tema de la hoja, theme
```

```python
import re

# Hacemos la función que va a clasificar qué tipo de contenido es el de la celda mediante expr regulares
def clasificar_celda(cell):
    valor = cell.value

    if valor is None:
        return

    texto = str(valor).strip()

    if texto.upper() in ["LUNES", "MARTES", "MIÉRCOLES", "JUEVES", "VIERNES"]:
        return ("DIA", texto)

    if re.match(r"^\d{1,2}:\d{2}$", texto):
        return ("HORA", texto)

    if texto.upper() is "HORA":
        return ("comienzo_tabla")
    
```
